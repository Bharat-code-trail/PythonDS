### *1) How does the number of hours studied per week correlate with the final exam score?*

```python
cor = df['Hours_Studied'].corr(df['Exam_Score'])
print(cor)
```

```python
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.xlabel('Hours_Studied')
plt.ylabel('Exam_Score')
plt.title('Correlation between Hours studied and Exam Score')
plt.show()
```

### *2) Is there a significant relationship between class attendance percentage and exam scores?*

```python
sns.lineplot(x='Attendance', y='Exam_Score', data=df)
plt.xlabel('Attendance')
plt.ylabel('Exam_Score')
plt.title('Relationship between Attendance and Exam Score')
plt.show()
```

### *3) Does the average number of sleep hours per night influence exam performance?*

```python
sns.lineplot(x='Sleep_Hours', y='Exam_Score', data=df)
plt.xlabel('Sleep Hours')
plt.ylabel('Exam Score')
plt.title('Relationship between Sleep hours and Exam score')
plt.show()
```

### *4) How does family income and parental education level impact the student's exam scores?*

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Family_Income', y='Exam_Score', data=df)
sns.barplot(x='Parental_Education_Level', y='Exam_Score', data=df)
plt.xlabel('Family Income/Parental Education Level')
plt.ylabel('Exam Score')
plt.title('Impact of Family Income/Parental Education Level on Exam Scores')
```

### *5) Do private school students perform better in exams than public school students?*

```python
df['School_Type'].value_counts()
```

```python
sns.barplot(x='School_Type', y='Exam_Score', data=df, hue='School_Type')
plt.xlabel('School Type')
plt.ylabel('Exam Score')
plt.title('Relationship between School Types and Exam Score')
plt.show()
```
