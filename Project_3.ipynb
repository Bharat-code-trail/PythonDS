### 1) What is the overall distribution of sentiments (positive, negative, neutral) across the posts?


```python
df['Sentiment'].value_counts()
```

```python
s_c = ['Positive', 'Negitive', 'Neutral']
f_c = df['Sentiment'][df['Sentiment'].isin(s_c)].value_counts()
print(f_c)
```

```python
sentiment_counts = df['Sentiment'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=sentiment_counts.index[:10], y=sentiment_counts.values[:10], palette="coolwarm")
plt.xticks(rotation=45)
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.title("Top 10 Sentiment Distribution")
plt.show()
```
